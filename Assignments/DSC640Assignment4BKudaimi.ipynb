```python
#Importing the necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#Importing and viewing the data
birth = pd.read_csv('birth-rate.csv')
crime = pd.read_csv('crimerates-by-state-2005.csv')
TV = pd.read_csv('tv_sizes.txt', sep = '\t')

print(birth.head())
print(crime.head())
print(TV.head())
```

```python
#Constructing the scatter plot
#I will plot the annual birth rate in the United States 

years = birth[birth['Country'] == 'United States'].iloc[0, 1:].index
rates = birth[birth['Country'] == 'United States'].iloc[0, 1:].values

#Generating the scatter plot
plt.scatter(years, rates)
plt.xticks(years[::2], rotation = 90)
plt.xlabel('Year')
plt.ylabel('Birth Rate (%)')
plt.title('Python - scatter plot \n Annual US Birth Rate')
plt.show()
```

```python
#Constructing the bubble chart
#I will plot how robbery changes with increasing aggravated assault for each US state, using population as the bubble size

#I will remove the national average and Washington DC since the US population is much 
#greater than its constituent states and Washington DC is not a state.
crime = crime.drop(crime[crime['state'] == 'United States'].index)
crime = crime.drop(crime[crime['state'] == 'District of Columbia'].index)

#Building the bubble chart
sns.scatterplot(x = 'aggravated_assault', y = 'murder', size = 'population', sizes = (20, 230), alpha = 0.7, data = crime)
plt.legend(loc = 'lower right', title = 'Population (e7)', bbox_to_anchor = (1.28, 0.55))
plt.xlabel('Aggravated Assault Rate')
plt.ylabel('Murder Rate')
plt.title('Python - bubble chart \n How US state murder rates change with assault rates')
plt.show()
```

```python
#Constructing the density plot
#I will show the probabiliy density of TV sizes from 2001 to 2009

sns.kdeplot(TV['size'], alpha = 0.7, fill = True)
plt.xlabel('TV size (inches)')
plt.title('Python - density plot \n Probability density of TV size distribution (2001-2009)')
plt.show()
```
