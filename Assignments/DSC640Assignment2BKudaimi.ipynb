```python
#Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

population = pd.read_excel('world-population.xlsm')
population.head()
```

```python
#Constructing the line chart
#I will plot the world population over time
year = population['Year']
pop = population['Population']
plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('World Population')
plt.title('Python - line chart \n World population (1960-2010)')
plt.show()
```

```python
#Constructing the step chart
#A step chart is similar to a line chart, but increases and decreases between points are represented in a step fashion
#I will plot the world population over time

plt.step(year, pop)
plt.xlabel('Year')
plt.ylabel('World Population')
plt.title('Python - step chart \n World population (1960-2010)')
plt.show()
```
